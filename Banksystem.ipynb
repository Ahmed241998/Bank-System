```python
# Creation of class
class Banksystem :
    # Intitiation function of the class
    def __init__(self,name,age,gender,balance) :
        # Class attribute that contains the name of the customer
        self.name=name
        # Class attribute that contains age of the customer 
        self.age=age
        # Class attribute that contain gender of the customer
        self.gender=gender
        # Class attribute that contain the balance of the customer
        self.balance=balance
    # Class method that will add the deposite amount to the balance
    def deposite(self,deposite) :
        self.balance += deposite
    # Class method that will substract the withdrawn amount from the balance
    def withdraw (self,withdraw) :
        # Check if the balance if enough for the withdrawn amount
        if withdraw <= self.balance :
            self.balance -= withdraw
        else :
            print("Not Enough Money")
    # Class method that display the balance
    def view_balance (self) :
        return self.balance
# Class CIB that inherit from Banksystem
class CIB(Banksystem) :
    # Class Method that will calculate how much the customer will pay for the loan based on 10% profit
    def loan_application (self,amount,duration) :
        print(f'Your Montly payrate for this loan {amount*1.1/(duration*12)}')
# Class QNB that inherit from Banksystem
class QNB(Banksystem) :
    # Class Method that will calculate how much the customer will pay for the loan based on 20% profit
    def loan_application (self,amount,duration) :
        print(f'Your Montly payrate for this loan {amount*1.2/(duration*12)}')
```

```python
def userinput_handling() :
    """
    Ask customer to enter his name, age, gender and balance
    Retuen 
    (str) name : The name of customer
    (int) age : the age of the customer
    (str) gender : the gender of the customer
    (int) balance : the balance of the customer 
    """
    # Use a while loop to handle invalid inputs and ask customer for correct selection
    while True :
        name = input("Enter Your Name : ").strip().title()
        if name.isalpha() == True :
            break
        else :
            print("Enter Your Name Correctly")
    # Use a while loop to handle invalid inputs and ask user for correct selection
    while True :
        try :
            age = input("Enter Your Age : ").strip()
            if 18<= int(age) <= 100 and age.isnumeric() == True :
                break
            else :
                print("PLease Enter Your Age Correctly")
        except :
            print("PLease Enter Your Age Correctly")
    # Use a while loop to handle invalid inputs and ask user for correct selection
    while True :
        gender = input ("Enter Your Gender : ").strip().title()
        if gender == "Male" or gender =="Female" :
            break 
        else :
            print("Enter Your Gender Correctly")
    # Use a while loop to handle invalid inputs and ask user for correct selection
    while True : 
        balance = input("Enter Your Balance : ")
        if balance.isnumeric() == True :
            balance = int(balance) 
            break
        else :
            print("Please Enter Your Balance Correctly")
    return name,age,gender,balance
```

```python
def process_handling() :
    """
    Ask customer to choose the number of the process he wants to do
    Returns :
    (int) process : the number of the process he wants to do
    """
    print("1.Deposite Money\n2.Withdraw Money\n3.View Balance\n4.Loan Services")
    # Use a while loop to handle invalid inputs and ask user for correct selection
    while True :
        try :
            process = int(input("Enter the number of your process : "))
            if 1<=process<=4 :
                break
        except:
            print("Enter Correct Choice")
    return process
```

```python
def main_process(name,age,gender,balance,process,bank): 
    """
    Act the process that the customer has chosen to do 
    """
    if process == 1 :
        while True :
            try :
                # Ask user to enter the amount he wants to deposite
                deposite=float(input("Enter The amount You Want To Be Added To Your Account : "))
                if 0 <deposite<= 20000 :
                    break
                else :
                    print("Please Enter Correct Number")
            except :
                print("Please Enter Correct Number")
        # Call Method to add the deposited amount to the balance
        bank.deposite(deposite)
    if process == 2 :
        while True :
            try :
                # Ask user to enter the amount he wants to withdraw
                withdraw=float(input("Enter The amount You Want To Withdraw From Your Acount : "))
                if 0 <withdraw<= 20000 :
                    break
                else :
                    print("Please Enter Correct Number")
            except :
                print("Please Enter Correct Number")
        # Call Method to Substract the withdrawn amount from the balance
        bank.withdraw(withdraw)
    if process == 3 :
        # call method to display the balance
        print(F"Your Balance is {bank.view_balance()}")
    if process == 4 :
        print("1.CIB\n2.QNB")
        # Use a while loop to handle invalid inputs and ask customer for correct selection
        while True :
            try :
                bank = int(input("Choose Your Bank "))
                if bank == 1 or bank == 2 :
                    break
                else :
                    print("Choose Your Bank Correctly")
            except :
                print("Choose Your Bank Correctly")
        if bank == 1 :
            # Use a while loop to handle invalid inputs and ask customer for correct selection
            while True :
                try :
                    # Ask customer to enter the amount he wants it as a loan
                    amount = int (input("Enter The amount of money You want to get it as a loan : "))
                    duration = int(input("Enter the Duration You want it"))
                    if amount < 1000000 :
                        break
                except :
                    print("Enter Correct Choice")
            bank1=CIB(name,age,gender,balance)
            bank1.loan_application(amount,duration)
        if bank == 2 :
            # Use a while loop to handle invalid inputs and ask customer for correct selection
            while True :
                try :
                    # Ask customer to enter the amount he wants it as a loan
                    amount = int (input("Enter The amount of money You want to get it as a loan : "))
                    duration = int(input("Enter the Duration You want it"))
                    if amount < 2000000 :
                        break
                except :
                    print("Enter Correct Choice")
            bank1=QNB(name,age,gender,balance)
            bank1.loan_application(amount,duration)
```

```python
name,age,gender,balance=userinput_handling()
process=process_handling()
bank=Banksystem(name,age,gender,balance)
main_process(name,age,gender,balance,process,bank)
while True :
    # check if customer wants to do another process
    user=input("Do You want another process [y/n] : ").strip().lower() 
    if user== "y" :
        process=process_handling()
        main_process(name,age,gender,balance,process,bank)
    elif user=="n" :
        break
    else :
        print("Please Enter Valid Input")
```
